```python
import pandas as pd
import numpy as np
```

```python
from tensorflow import keras
```

```python
from keras.preprocessing.text import one_hot
```

```python
# Sentence 
sent=['the glass od milk','the glass of juice',
      'the cup of tea','i m a good boy','i m a good developper',
      'understand the Remaining of words','your video are good']
```

```python
sent
```

```python
# Vocabulary Size
voc_size=10000
```

```python
onehot_repr=[one_hot(word,voc_size) for word in sent]
print(onehot_repr)
```

```python
from keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
from tensorflow.keras.models import Sequential
```

```python
sent_length=8
embeded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
```

```python
print(embeded_docs)
```

```python
# Select Feature means how many many u want to take 
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embeded_docs))
```
